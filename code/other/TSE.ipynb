```pyspark
df_final = spark.sql('''
    SELECT
        cpf_cnpj,
        nome_cpf_cnpj,
        data,
        tipo_entrada,
        destino,
        nome_candidato,
        cargo_candidato,
        partido,
        uf,
        descricao,
        valor
    FROM
        despesas_candidatos
    UNION ALL
    SELECT
        cpf_cnpj,
        nome_cpf_cnpj,
        data,
        tipo_entrada,
        destino,
        nome_candidato,
        cargo_candidato,
        partido,
        uf,
        descricao,
        valor
    FROM
        receitas_candidatos
    UNION ALL
    SELECT
        cpf_cnpj,
        nome_cpf_cnpj,
        data,
        tipo_entrada,
        destino,
        nome_candidato,
        cargo_candidato,
        partido,
        uf,
        descricao,
        valor
    FROM
        despesas_partidos
    UNION ALL
    SELECT
        cpf_cnpj,
        nome_cpf_cnpj,
        data,
        tipo_entrada,
        destino,
        nome_candidato,
        cargo_candidato,
        partido,
        uf,
        descricao,
        valor
    FROM
        receitas_partidos
    UNION ALL
    SELECT
        cpf_cnpj,
        nome_cpf_cnpj,
        data,
        tipo_entrada,
        destino,
        nome_candidato,
        cargo_candidato,
        partido,
        uf,
        descricao,
        valor
    FROM
        despesas_comite
    UNION ALL
    SELECT
        cpf_cnpj,
        nome_cpf_cnpj,
        data,
        tipo_entrada,
        destino,
        nome_candidato,
        cargo_candidato,
        partido,
        uf,
        descricao,
        valor
    FROM
        receitas_comite
''')
df_final.createOrReplaceTempView('final')
df_final.show(vertical=True, truncate=False)
```

```pyspark
df_agrupado = spark.sql('''
    SELECT
        cpf_cnpj,
        count(cpf_cnpj) AS quantidade,
        SUM(valor) AS valor_total
    FROM
        final
    GROUP BY 
        cpf_cnpj
''')
df_agrupado.show(vertical=True, truncate=False)
```
